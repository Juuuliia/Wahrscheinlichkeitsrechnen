### Kombinatorik

```python
# Reihenfolge wird berücksichtigt
# Element darf mehrmals gewählt werden 

def Variationen_mit_Wiederholung():
    # n = Anzahl Elemente in der Grundmenge
    n = 3
    # k = Anzahl der Ziehungen
    k = 3
    print(n ** k)
    
Variationen_mit_Wiederholung()
```

```python
# Reihenfolge wird berücksichtigt
# Element darf nur einmal gewählt werden

def Variationen_ohne_Wiederholung():
    import math
    # n = Anzahl Elemente in der Grundmenge
    n = 2
    # k = Anzahl der Ziehungen
    k = 5
    result = (math.factorial(n) / (math.factorial(k) * math.factorial(n - k))) * math.factorial(k)
    print(result)
    
Variationen_ohne_Wiederholung()
```

```python
# Reihenfolge spielt keine Rolle
def Kombinationen_mit_Wiederholung():
    import math
    # n = Anzahl Elemente in der Grundmenge
    n = 2
    # k = Anzahl der Ziehungen
    k = 5
    
    temp = n + k - 1
    result = math.factorial(temp) / (math.factorial(k) * math.factorial(temp - k))
    print(result)
    
Kombinationen_mit_Wiederholung()
```

```python
# Reihenfolge spielt keine Rolle
def Kombinationen_ohne_Wiederholung():
    import math
    # n = Anzahl Elemente in der Grundmenge
    n = 10
    # k = Anzahl der Ziehungen
    k = 4
    result = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    print(result)
          
Kombinationen_ohne_Wiederholung()
```

```python
# Fakultät ausrechnen ohne Wiederholung (alles kommt nur ein mal vor, alles ist verschieden)
def Permutation_ohne_Wiederholung():
    import math
    # n = Anzahl Elemente in der Grundmenge
    n = 6
    result = math.factorial(n)
    print(result)

Permutation_ohne_Wiederholung()
```

```python
# Fakultät ausrechnen mit Wiederholung (von etwas hat es mehrere Sachen)
def Permutation_mit_Wiederholung():
    import math
    # n = Anzahl Elemente in der Grundmenge
    n = 5
    # k = Anzahl gleiche Elemente (jede Gruppe gleicher Elemente einzeln definieren) Je nach Bedarf k3, k4 
    # usw. ergänzen (AUCH IN FORMEL!)
    k1 = 2
   
    result = math.factorial(n) / (math.factorial(k1))
    print(result)
Permutation_mit_Wiederholung()
```

### Punktschätzer / Konfidenzintervalle

```python
def Punktschätzer_Mittelwert():
    print("Punktschätzer Mittelwert")
    # https://spaces.technik.fhnw.ch/multimediathek/file/parameterschatzung-skript
    import numpy as np
    
    data = [24, 34, 32, 36, 38, 32, 28]
    
    print(round(np.mean(data), 3))
    
Punktschätzer_Mittelwert()
```

```python
def Punktschätzer_Standardabweichung():
    print("Punktschätzer Standardabweichung")
    # https://spaces.technik.fhnw.ch/multimediathek/file/parameterschatzung-skript
    import numpy as np
    
    data = [24, 34, 32, 36, 38, 32, 28]
    
    print("Varianz:", round(np.var(data, ddof=1), 3))
    print("Standardabweichung:", round(np.std(data, ddof=1), 3))
    
Punktschätzer_Standardabweichung()
```

```python
def Konfidenzintervall_beidseitig_Twert():
    import scipy.stats as st
    
    # Stichprobengroesse
    n = 225
    
    # Freiheitsgrade 
    f = n - 1
    
    # Mittelwert
    mue = 600000
    
    # Standardabweichung
    std = 90000
    
    # konfi = Gesuchter Konfidenzintervall
    konfi = 0.9545
    alpha = (1 - konfi) / 2
    
    t_wert = st.t.ppf(1 - alpha, f)
    
    Untere_Grenze = mue - t_wert * (std / n ** 0.5)
    Obere_Grenze = mue + t_wert * (std / n ** 0.5)
    
    print(Untere_Grenze)
    print("Stichprobengroesse: {0}, Freiheitsgrade: {1}, Mue: {2}, Standardabweichung: {3}".format(n,f,mue,std))
    print("Konfidenzintervall: {0}, alpha: {1}, t-wert {2}".format(konfi,round(alpha,6),t_wert))
    print("Untere Grenze: {0},  Obere Grenze: {1},".format(Untere_Grenze, Obere_Grenze))
    
Konfidenzintervall_beidseitig_Twert()
```

```python
def Konfidenzintervall_beidseitig_zwert():
    import scipy.stats as st
    
    # Stichprobengroesse
    n = 225
    
    # Freiheitsgrade 
    f = n - 1
    
    # Mittelwert
    mue = 600000
    
    # Standardabweichung
    std = 90000
    
    # konfi = Gesuchter Konfidenzintervall
    konfi = 0.9545
    alpha = (1 - konfi) / 2
    
    z_wert = st.norm.ppf(1 - alpha)
    
    Untere_Grenze = mue - z_wert * (std / n ** 0.5)
    Obere_Grenze = mue + z_wert * (std / n ** 0.5)
    
    print(Untere_Grenze)
    print("Stichprobengroesse: {0}, Freiheitsgrade: {1}, Mue: {2}, Standardabweichung: {3}".format(n,f,mue,std))
    print("Konfidenzintervall: {0}, alpha: {1}, z-wert {2}".format(konfi,round(alpha,6),z_wert))
    print("Untere Grenze: {0},  Obere Grenze: {1},".format(Untere_Grenze, Obere_Grenze))
    
Konfidenzintervall_beidseitig_zwert()
```

### Exponentialverteilung

```python
def Exponentialverteilung():
    print("Exponentialverteilung")
    # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon
    from scipy.stats import expon
    
    # lamba entspricht der durchschnittlichen Wartezeit
    lamba = 10
    
    mean, var, skew, kurt = expon.stats(loc=0, scale=lamba, moments='mvsk')
    std = expon.std(loc=0, scale=lamba)
    print("Mittelwert:", mean, "Varianz:", var, "Standardabw.:", round(std, 2), "Skewness:", skew, "Kurtosis:", kurt)
    
    # x entspricht Beobachtungszeitraum
    x = 20
    
    print("Wahrscheinlichkeit genau x: PDF:", expon.pdf(x, loc=0, scale=lamba))
    print("Wahrscheinlichkeit bis und mit x: CDF:", expon.cdf(x, loc=0, scale=lamba))   

Exponentialverteilung()
```

### Poissonverteilung

```python
def Poissonverteilung():
    print("Poissonverteilung")
    # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html#scipy.stats.poisson
    from scipy.stats import poisson
    
    # Mu entspricht erwartete Anzahl Eintreten
    mu = 4.1
    
    mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
    std = poisson.std(mu, loc=0)
    print("Mittelwert:", mean, "Varianz:", var, "Standardabw.:", round(std, 2), "Skewness:", skew, "Kurtosis:", kurt)
    
    # x entspricht Anzahl Treffer
    x = 3
    
    print("Wahrscheinlichkeit genau k: PDF:", poisson.pmf(x, mu, loc=0))
    print("Wahrscheinlichkeit bis und mit k: CDF:", poisson.cdf(x, mu, loc=0))   

Poissonverteilung()
```
